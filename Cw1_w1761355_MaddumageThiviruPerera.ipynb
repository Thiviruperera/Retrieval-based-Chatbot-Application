# Compare and evaluate AI techniques

## Retrieval-based Model

Traditionally, chatbots that use retrieval-based models have relied on rule-based systems for answering questions. These chatbots have a set of predetermined responses that are linked to specific questions. Some advanced models of this are able to store information about the context of a conversation, generate multiple responses based on that context, evaluate the appropriateness of each response, and then select and output the response with the highest score.

Retrieval-based chatbots are well-suited for customer service applications because they can quickly and accurately provide responses to common questions. Additionally, these chatbots can be easily trained to handle a wide range of inputs and outputs, making them a practical choice when creating a customer service chatbot. Furthermore Retrieval-based chatbots do not need to have a thorough understanding of the user's intentions, which can make them more reliable and easier to develop compared to other types of chatbots. (Bhagwat, 2018)

When reviewing the advantages and strengths of retrieval-based chatbots, it becomes evident that this type of chatbot model is efficient because it can quickly choose a response from a pre-defined set rather than processing the input and producing a new response, which makes it fast and accurate. Furthermore retrieval-based chatbots are relatively easy to construct and maintain because they do not rely on complex machine learning models or large amounts of training data. They can also handle a large volume of requests concurrently without requiring significant computational resources. It can also provide information on a wide variety of subjects and can cover a large range of topics. (Chung, J.,2018)

By evaluating these, it is clear that retrieval-based chatbots can be useful in delivering customer service in circumstances like: Responding to frequently requested queries, providing basic information, and resolving straightforward concerns.

On the contrary, using a retrieval-based model for chatbots has significant shortcomings such as limited adaptability. Retrieval-based chatbots are constrained to the predefined responses, they are unable to create new responses or adapt to novel input. Retrieval-based chatbots may also have trouble maintaining context or comprehending the flow of the entire discussion because they are unable to save and retrieve data from previous exchanges. Additionally, because this model cant personalize it's responses to specific users or the context of the conversation, retrieval-based chatbots can come out as impersonal or uninvolved. (Tavakoli, 2018)

When assessing the retrieval-based model for chatbots, there are certain additional considerations to take into account, Retrieval-based chatbots need a pre-determined set of inputs and outputs, so the data they can access is just what has been given. Additionally, compared to other types of chatbots, these are generally easy to construct and do not require substantial training, therefore setup times are typically quicker. They can also swiftly choose a response from their pre-defined selection, making them generally quick to produce results. Finally, A retrieval-based chatbot's output is restricted to the pre-defined responses, hence it might not be able to produce fresh or inventive responses. However, it can provide quick and efficient responses to common questions and issues. 


## Generative Model

A generative chatbot is an open domain chatbot program that creates unique linguistic combinations as opposed to choosing from pre-defined responses. Generative model chatbots use artificial intelligence techniques, such as machine learning, to generate new responses based on the input they receive. These responses are designed to be similar to human responses, and the goal of these chatbots is to have a conversation with a human in a natural and engaging way. Some generative model chatbots learn patterns and structures from large datasets of human interaction and then use these patterns and structures to produce responses. Generative chatbots can be created using sequence-to-sequence (Seq2Seq) models, which are also used for machine translation. (Bhagwat, 2018)

Generative models can be used to create chatbots that can produce responses that are similar to human responses to customer inquiries. These chatbots can be trained on large collections of customer service conversation dataset enabling them to learn the language and style that is commonly used in these interactions. The chatbot can then be used to generate real-time customer support responses after being trained on this dataset. When a consumer messages the chatbot, the chatbot can leverage its understanding of the language and conversational patterns of customer service discussions to produce a response that resembles what a real customer support agent might provide.(Shah, 2020)

When discussing the advantages and strengths of using generative models for chatbots, we can mention that these models can generate responses that are more varied and unpredictable than those generated by rule-based chatbots, as well as responses that are more engaging and human-like, which may make the user feel as though the conversation is more natural and realistic. Additionally, it can handle a wide variety of input and can reply effectively to inquiries and prompts that they were not specifically trained on. moreover, it can learn from the training data, which allows it to develop its responses over time. (Chung, J.,2018)

By weighing these, it is evident that generative chatbots can be helpful in customer service for responding to frequent consumer inquiries, giving individualized recommendations, and elevating complex situations to human agents.

But the use of a generative model for chatbots, on the other hand, has potential disadvantages and weaknesses like the possibility of producing nonsensical or inappropriate responses because of the probabilistic method of their creation. Also requiring a significant amount of training data in order to generate reliable and accurate results. Furthermore, as they need to process the input and produce a response, they could be slower than rule-based chatbots. Also, it might not be able to keep the same "personality" or tone throughout the discussion. (Tavakoli, 2018)

In conclusion, a generative chatbot needs a large dataset of real-world conversations for training, and the availability and caliber of this data might affect the design and functionality of the chatbot. It can take some time to develop a generative chatbot because it requires gathering and preparing training data, training a machine learning model, and adjusting the chatbot's performance. The complexity of the task, the quantity of data and computational resources available, as well as the quality of the training data and the effectiveness of the machine learning model, all affect how quickly generative chatbots can provide results.


## Template-based Model

A template-based chatbot is a type of chatbot that uses a predefined set of templates to generate responses to user input. This method is straightforward and can be useful for handling basic customer inquiries or providing information on a specific set of topics.

A series of templates that address a variety of probable user enquiries would need to be defined in order to develop a chatbot that uses templates. Every template would include placeholders for dynamic elements like usernames or specific information about a product or service. The first time a user interacts with the chatbot, it will try to match their input to one of the templates. If a match is found, the chatbot can then construct a response by inserting the necessary values in the template's placeholders. (Bhattacharya et al., 2016)

Template-based model is a response generated based model and it can be used in technologies like AIML, Pattern Matching, Fact Generator, Initiator Model, Storybot Model and BoW Model. Although they are less flexible and may not be able to handle more sophisticated or open-ended inputs, they are excellent for managing a small number of very specific queries or inputs. They are frequently used as the user interface for more complex chatbots that employ cutting-edge techniques like machine learning or natural language processing. (Nuruzzaman and Hussain, 2018)

However, the responses produced in template-based models are less diverse and predictable when compared with retrieval-based models. Additionally, it is less adaptable than a retrieval model, which can handle a wider range of inputs but requires more training data and is more difficult to develop and maintain.

It is noticeable when examining the advantages of a template-based model that because they rely on a fixed set of pre-defined responses, template-based models are relatively easy to construct and maintain. Additionally, these chatbots are simple for consumers to communicate with because they respond to specific requests or inputs in a straightforward and predictable manner. Furthermore, since they do not need to analyze or interpret the user's input in order to respond, template-based chatbots can respond to users quickly. (Kaur et al., 2017)

Template-based customer service chatbots can be integrated into a company's website or social media channels to provide quick answers to customer questions without the need for a human customer service representative. These chatbots can also be used to handle simple tasks such as checking the status of an order or scheduling an appointment. These are some examples how this technique is being used in customer service.

In summary, template-based models for chatbots can produce results quickly, are easy to set up, and do not require a large amount of data.  They can only process a limited number of queries or inputs and rely on pre-defined responses kept in a database. The care and consideration used to create the templates will determine the quality of the responses. The pre-defined responses retrieved from the database based on the user's input are the output of a template-based chatbot. However, template-based chatbots may not be able to handle more complex or open-ended inputs and may provide predictable or repetitive responses. (Kaur et al., 2017)

For this system, as data we will use a .JSON (JavaScript Object Notation) file to define specific keywords that the chatbot will recognize and use to determine how to respond. This system doesn’t require to download any huge datasets because it’s a straightforward chatbot. Here we shall only make use of data that we have created since it’s a closed domain retrieval-based chatbot.

The JSON file includes intents, for each intent there is a tag, question patterns, responses, and a context. This file generates a number of messages that the user is probably going to input and maps them to a group of appropriate answers. Each dictionary in the file has a tag that identifies the group to which each message belongs. With the use of this data, we will train a neural network to identify a sentence of words and classify it as one of the tags in our file. Then a response will be selected from those groups and be presented it to the user. The more tags, responses, and patterns we provide to the chatbot, the more advanced and sophisticated it will become.

The data needs to be preprocessed in a specific way before it can be used in this system’s model. Extracting the data including patterns of which class or tag they belong to from the JSON file and storing all the unique words from those patterns in some blank lists is the first step.

Then after extracting all the data by looping through the JSON file we have to convert each pattern into a list of words using nltk.word_tokenizer instead of having them as strings. Then add the patterns and its associated tag into the relevant lists.

Then the data must be subjected to the word stemming process to reduce the vocabulary of the model and attempt to find more general meaning behind sentences. Here, the goal is to find the root of the word. As an example word “let’s” stem will be “let” and the word “running” would have the stem of “run”.

After creating a unique list of stemmed words, the final step of data preprocessing is to create a bag of words representation. As it is well known, numerical input is necessary for neural networks and machine learning algorithms. Therefore, the list of strings won't work.

We'll use a list with the length of the number of items in the model's vocabulary to represent each sentence. A word from our vocabulary will be represented by each slot in the list. If a word is listed at position 1, it means that it is present in our sentence; if it is listed at place 0, it is not. We refer to this as a "bag of words" because we don't know the exact order in which the words appear in the phrase; instead, we only know that they are there in the vocabulary model.

# Importing modules and loading data

You can download the intents.json file from here - https://drive.google.com/file/d/16mBw-UejDdN1x4A-3YAvY3mr1jqAef7r/view?usp=sharing

```python
import nltk
from nltk.stem.lancaster import LancasterStemmer
stemmer = LancasterStemmer()
nltk.download('punkt')

import numpy as np
import tflearn
import tensorflow as tf
import random as rd
import json
import matplotlib.pyplot as plt
import pickle

with open("intents.json") as file:    #Loading data from the json file
    data = json.load(file)
```

# Extracting Data

```python

    words = []       #Creating blank lists to store unique values
    labels = []
    docs_x = []
    docs_y = []

    for intent in data["intents"]:
        for pattern in intent["patterns"]:       
            wrds = nltk.word_tokenize(pattern)  #Converting strings into a list of words
            words.extend(wrds)
            docs_x.append(wrds)            #Adding each pattern into docs_x list
            docs_y.append(intent["tag"])   #Adding its associated tag into docs_y list

        if intent["tag"] not in labels:
            labels.append(intent["tag"])
```

# Word Stemming

```python
words = [stemmer.stem(w.lower()) for w in words if w != "?"]  #Removing duplicate elements while stemming the words
words = sorted(list(set(words)))    #Creating a unique list of stemmed words

labels = sorted(labels)    
```

# Bag of words

```python
training = []
output = []

out_empty = [0 for _ in range(len(labels))]

#Conveting the strings to numerals, One hot encoding

for x, doc in enumerate(docs_x):
    bag = []

    wrds = [stemmer.stem(w.lower()) for w in doc]

    for w in words:
        if w in wrds:
            bag.append(1)
        else:
            bag.append(0)

    output_row = out_empty[:]
    output_row[labels.index(docs_y[x])] = 1

    training.append(bag)
    output.append(output_row)
```

```python
training = np.array(training)    #Converting the training data and output to numpy arrays
output = np.array(output) 
```

# Model Development

```python
#Building the Neural Network

tf.compat.v1.reset_default_graph()

net = tflearn.input_data(shape=[None, len(training[0])])
net = tflearn.fully_connected(net, 8)
net = tflearn.fully_connected(net, 8)
net = tflearn.fully_connected(net, len(output[0]), activation="softmax")
net = tflearn.regression(net)

model = tflearn.DNN(net)
```

# Training and Saving the Model

```python
# Define model and setup tensorboard
model = tflearn.DNN(net, tensorboard_dir='tflearn_logs')

# Start training 
history = model.fit(training, output, n_epoch=1000, batch_size=8, show_metric=True)
model.save("model.tflearn")
```

# Creating Predictions

```python
def bag_of_words(s, words):
    bag = [0 for _ in range(len(words))]

    s_words = nltk.word_tokenize(s)
    s_words = [stemmer.stem(word.lower()) for word in s_words]

    for se in s_words:
        for i, w in enumerate(words):
            if w == se:
                bag[i] = 1
            
    return np.array(bag)


def chat():
    print("Start talking with the bot (type quit to stop)!")
    while True:
        inp = input("You: ")
        if inp.lower() == "quit":
            break

        results = model.predict([bag_of_words(inp, words)])[0]  #Probability of giving correct response
        results_index = np.argmax(results)   #Choosing the largest number from the probability
        tag = labels[results_index]
        
        if results[results_index] > 0.7:
            for tg in data["intents"]:
                if tg['tag'] == tag:
                    responses = tg['responses']

            print(rd.choice(responses))
        else:
            print("I don't quite understand. Try again or ask a different quesion")
        
chat() 
```
